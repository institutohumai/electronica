# Introduccion - Teoria


## ¿Que es una API REST?

API son las en ingles para: Application Programming Interface, al entender el porque que cada palabra, entender que es una API es mas sencillo:

- **Application (Aplicación):** Se refiere a cualquier programa o software diseñado para realizar un conjunto de tareas o funciones. Piensa en cualquier aplicación en tu teléfono o computadora; todas están diseñadas para ayudarte a lograr algo, saber el clima, ver los mapas, enviar mensajes, editar fotos, o navegar por internet.

- **Programming (Programación):** La programación es lo que permite desarrollar las instrucciones que las computadoras y aplicaciones siguen para realizar sus tareas, permite a los desarrolladores usar funcionalidades ya programadas sin tener que crearlas desde cero.

- **Interface (Interfaz):** Una interfaz es un punto de interacción o comunicación. En el contexto de las APIs, se refiere al conjunto de reglas y protocolos que permiten que una aplicación interactúe con otra aplicación, servicio o plataforma. Es similar a una interfaz grafica de usuario, que es donde interactúas con una aplicación, pero en este caso, la "interfaz" es para la comunicación entre aplicaciones, no entre humanos y software.

Luego "REST" quiere decir Representational State Transfer el cual el un estilo arquitectonico para crear APIs web usando HTTP. En la web asi como las API rest existen otros tipos de API como las SOAP o GraphQL, pero nosotros nos enfocaremos en las API REST.

Entonces de manera resumida podemos definir:

**Una API es un conjunto de reglas y especificaciones que las aplicaciones usan para comunicarse entre sí. Permite que una aplicación "solicite" funciones o datos de otra aplicación, servicio o sistema operativo de manera estandarizada, facilitando la integración y la interoperabilidad entre diferentes softwares.**

- La aplicacion o entidad que solicita la informacion a la API se le conoce como CLIENTE.
- La aplicacion o entidad que procesa la solicitud y devuelve la informacion se le conoce como SERVIDOR.
- La herramienta que actua como un puente de comunicacion entre el CLIENTE y el SERVIDOR es la API.


## Conceptos importantes

### HTTP (Hyper Text Transfer Protocol)

Es un protocolo que permite usar las URL para realizar un pedido a un servidor y luego obtener una respuesta, el protocolo establece una serie de METODOS que permite al cliente comunicarle al servidor que ACCION quiere realizar y tambien establece una serie de CODIGOS DE ESTADO que permite al servidor comunicarle al cliente que sucedio con su solicitud.



### Metodos o Acciones HTTP principales (para indicarle al servidor que hacer)

### Codigos de Estado HTTP principales (para entender que sucedio con la solicitud)

| Código | Estado | Descripción |
|--------|--------|-------------|
| 200    | OK | La solicitud ha tenido éxito y la respuesta contiene los datos solicitados. |
| 201    | Created | La solicitud ha sido cumplida y ha resultado en la creación de un nuevo recurso. |
| 400    | Bad Request | La solicitud no se puede procesar debido a un error del cliente, como una solicitud malformada. |
| 401    | Unauthorized | La solicitud requiere autenticación del usuario y ha sido denegada por falta de credenciales válidas. |
| 403    | Forbidden | El servidor ha entendido la solicitud, pero se niega a autorizarla. |
| 404    | Not Found | El servidor no ha encontrado nada que coincida con el URL de la solicitud. |
| 500    | Internal Server Error | El servidor ha encontrado una situación que no sabe cómo manejarla. |

## 4 puntos para identificar rapidamente una API REST


1. Son URLs que utilizan el protocolo HTTP para realizar solicitudes y obtener respuestas (a traves de los metodos GET, POST, PUT, DELETE).
2. El formato de la informacion enviada o recibida casi siempre es JSON, existiendo la posibilidad de encontrarse con alguna que use XML.
3. Suelen tener sistemas de autenticacion, que comunmente llamamos "token", "api token", "api secret", "client secret" para controlar quien puede acceder a la informacion.
4. En el mejor de los casos vienen acompañadas de documentacion que explica como usarla.

## Composicion de la URL de una API REST

Antes de ponernos manos a la obra, debemos entender como se puede componer una URL.

---
> 💡 ***Sabias que?***
>
> *Las URLs de una API tambien se suelen llamar Endpoints o Rutas.*
---

Consideremos el siguiente endpoint de ejemplo:

```
https://www.api-ejemplo.com/usuarios/42/perfil?tema=oscuro&notificaciones=email
```

Podemos observar:

- **Recurso**: El recurso es `/usuarios/42/perfil`. Indica que la solicitud está dirigida a acceder al perfil de un usuario específico en el servidor.

  - **Parámetros de Ruta**: En la parte `/usuarios/42`, `42` es un parámetro de ruta. Los parámetros de ruta son parte de la URL que a menudo identifican un recurso específico dentro de la colección, en este caso, el `ID` del usuario en la sección de usuarios. La ruta `/usuarios/42` indica que estamos accediendo al usuario con el ID `42`.

- **Parámetros de Cadena de Consulta**: Después del signo de interrogación `?` y separados por el signo de Ampersand `&`, tenemos:
  
  - `tema=oscuro`: Este es un parámetro de consulta, donde `tema` es la clave y `oscuro` es el valor. Indica que el usuario prefiere un tema oscuro para la interfaz.
  
  - `notificaciones=email`: Otro parámetro de consulta, donde `notificaciones` es la clave y `email` es el valor, indicando que el usuario prefiere recibir notificaciones por correo electrónico.

En este ejemplo, la URL combina parámetros de ruta (para especificar el ID del usuario cuyo perfil se está accediendo) y parámetros de cadena de consulta (para indicar preferencias adicionales como el tema de la interfaz y el método de notificación preferido).

---
> 💡 ***Sabias que?***
>
> *Los parametros de consulta son tambien conocidos como parametros opcionales o en ingles como Query Parameters, estos siempre van despues del `?` y sirven para decirle al servidor como filtrar o procesar la informacion del recurso.*

## Con Python


Para consumir APIs usando Python, una de las maneras más comunes y sencillas es utilizar la biblioteca requests. Esta biblioteca te permite enviar solicitudes HTTP y procesar sus respuestas de manera muy sencilla.

Primero asegurate de tenerla instalada:
```
pip install requests
```

````python
import requests
````

### solicitud tipo GET simple

````python
# La URL de la API que quieres consumir
url = "https://www.cultura.gob.ar/api/v2.0/museos/"

# Realiza una solicitud GET a la URL
respuesta = requests.get(url)

if respuesta.status_code == 200:
    datos = respuesta.json()
else:
    print("La solicitud a fallado")
````

````python
[resultado["nombre"] for resultado in datos['results']]
````

### Solicitud tipo GET con parametros de consulta y de ruta

#### Opcion 1 (parametros en la url)

````python
# La URL base de la API
url = "https://restcountries.com/v3.1/name/argentina?fields=borders"

# Realiza la solicitud GET incluyendo los parámetros de la consulta
respuesta = requests.get(url)

if respuesta.status_code == 200:
    print("Resultados:", respuesta.json())
else:
    print("Error en la solicitud:", respuesta.status_code)
````

#### Opcion 2 (Mas Pythonica)

````python
# La URL base de la API
url = "https://restcountries.com/v3.1/name/argentina"

# Los parámetros de la consulta
params = {
    "fields": "borders"
}

# Realiza la solicitud GET incluyendo los parámetros de la consulta
respuesta = requests.get(url, params=params)

if respuesta.status_code == 200:
    print("Resultados:", respuesta.json())
else:
    print("Error en la solicitud:", respuesta.status_code)
````

````python
respuesta.json()[0]["borders"]
````

Lo bueno de usar Python es que podemos procesar los datos como queramos luego de obtener la respuesta:

````python
# conseguir las banderas de los paises limitrofes de Argentina
fronteras_argentina = respuesta.json()[0]["borders"]
for frontera in fronteras_argentina:
    respuesta_temporal = requests.get(f"https://restcountries.com/v3.1/alpha/{frontera}?fields=flags")
    if respuesta_temporal.status_code == 200:
        print(respuesta_temporal.json()['flags']['png'])
````

### Solicitud POST y como autorizar

````python
url = "https://api.openqr.io/api/v1/qr-codes"

payload = {
  "name": "test humai",  # cambia esto si quieres
  "type": "url",
  "data": {
    "url": "https://humai.com.ar"
  },
  "design": {
    "module": "roundness",
    "finder": "eye-shaped",
    "finder_dot": "eye-shaped",
    "frame": "border-around-text-bottom",
    "frame_text": "HUMAI"
  }
}
headers = {
  'Authorization': 'Coloca tu Bearer api_token aqui',
}

response = requests.post(url, headers=headers, json=payload)
````

````python
if response.status_code == 201:
    qr_info = response.json()['data']
````
